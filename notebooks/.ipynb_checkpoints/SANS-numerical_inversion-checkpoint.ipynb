```python
# In this notebook the pair distance distribution function is extracted from 
# the small-angle scattering data (here magnetic SANS data) of a nanoparticle sample
# by a regularized numerical inversion (i.e., an indirect Fourier transform IFT).
# The results using basically the identical code are published in: 
# "Dipolar-coupled moment correlations in clusters of magnetic nanoparticles"
# https://doi.org/10.1103/PhysRevB.98.224420

#==============================================================================
# import of used packages
from numpy.linalg import slogdet
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import nnls, lsq_linear
#==============================================================================
# import of data

q,Iq,sd=np.genfromtxt("../data/NC-SANS_data.txt",unpack=True,skip_header=0,skip_footer=0,usecols=(0,1,2))
  
Nq=len(q)

sd_sq   = pow(sd,2)       
sdnorm  = 1/sd 
sdnorm = Nq * sdnorm/sdnorm.sum()
sdmat = np.diag(sdnorm)
sdcol  = sdnorm[:,np.newaxis]            

I = Iq.dot(sdmat) # normalize data by sigma

N=100 # N is number of r values in determined pair distance distribution function

#==============================================================================
# generate list for regularization parameter a and diameter D

ND=71
DList=np.linspace(10,80,ND) # Diameter list

Na=100
aList=np.logspace(0,7,Na) # regularization parameter

#==============================================================================
# Plot data

plt.errorbar(q,Iq,sd,color='r',marker='.',ls='None',label='Observed')
plt.xscale('log')
plt.yscale('log')
plt.show()
print('Figure 1: data')   

#==============================================================================
# regularization matrices

L=0.5*(-1*np.eye(N,k=-1)+2*np.eye(N,k=0)-1*np.eye(N,k=1))
L[0,1] = 0
L[N-1,N-2] = 0

#==============================================================================
# Definition of function to generate the system matrix

def Aij(q,r,dr): # IFT
    y=4*np.pi*np.sin(q*r)/(q*r)*dr 
    return y

#==============================================================================
# Inversion of data (for varying D and a values the data is inversed, evidence determined)
evList1=[] 
evList=[] 
prList=[]

X = np.hstack([I, np.zeros(N)]) 
qmat=np.tile(q,(N,1)).T

for Di in range(len(DList)): # for loops could be replaced by numpy arrays to speed up calculation time
                             # but then the matrices would become huge 
    
    # defintion of r-List       
    r=np.linspace(0,DList[Di],N)
    dr=np.ones(N)*(r[1]-r[0])
    r[0]=+1.e-15
    
    rmat=np.tile(r,(Nq,1))
    drmat=np.tile(dr,(Nq,1))
              
    K=Aij(qmat,rmat,drmat) 
    K=np.asarray(K)
    
    B = (K.T).dot(K*pow(sdcol,2)) 
    
    for ai in range(len(aList)):
        # stacking of matrices            
        C = np.vstack([K*sdcol,np.sqrt(aList[ai])*L]) 
        
        # Fit: 
#        pr, resnorm = nnls(C,X) # least-square fit with non-negative constraint
        pr=np.dot(np.dot(np.linalg.inv(np.dot(C.T,C)),C.T),X) # Least-square fit without non-negative constraint
        
        # determine chisquare            
        Ifit=K.dot(pr)    
        chisq=(pow(Iq-Ifit,2)/sd_sq).sum()               

        # calculation of evidence 
        S=sum(pow(L.dot(pr),2))
        Lsq=pow(L,2)                   
        U=2*Lsq+B/aList[ai] 
        detsign,rlogdet=slogdet(U)
        G=-aList[ai]*S-0.5*chisq        
        evidence=(G-0.5*rlogdet) # log of evidence
            
        evList1.append(np.hstack([aList[ai],DList[Di],evidence]))
        
        evList.append(evidence)            
        prList.append(pr) 

# Search for maximal Diameter with highest evidence
MAX=evList.index(max(evList))

evList1=np.asarray(evList1)
DM=evList1[MAX,1]
prM=prList[MAX] # pair distance distribution function for which highest evidence was calculated

MAXD=np.where(DList == DM)

# maximal diameter is fixed to DM
# r-List is constructed for DM and pr's for all alpha values written in new list
# as well as corresponding evidence values

Nqfit=100
qfit=np.logspace(-2,0,Nqfit)
qfitmat=np.tile(qfit,(N,1)).T

r=np.linspace(0,DM,N)
dr=np.ones(N)*(r[1]-r[0])
r[0]=+1.e-15

rmat=np.tile(r,(Nqfit,1))
drmat=np.tile(dr,(Nqfit,1))

prList2=np.reshape(prList,(ND,Na,N))
prList22=np.reshape(prList2[MAXD],(Na,N))

evList2=np.reshape(evList,(ND,Na))
evList22=np.reshape(evList2[MAXD],(Na))
#==============================================================================
# Reconstruction of I(q)
        
K=Aij(qfitmat,rmat,drmat) 
K=np.asarray(K)

Iqfit=K.dot(prM) # fitting curve (reconstructed for prM) 

#==============================================================================
# Plots

plt.plot(r,prM)
plt.show()   
print('Figure 2: Pair distance distribution function prM.')
print('particle size = '+repr(DM)+' nm')

plt.figure()
for i in range(Na):
    plt.plot(r,prList22[i,:])
plt.show() 
print('Figure 3: all distributions.')

plt.errorbar(q,Iq,sd,color='r',marker='.',ls='None',label='Observed')
plt.loglog(qfit,Iqfit)
plt.show() 
print('Figure 4: Data (red points) and fit (blue line).')

plt.plot(aList,evList22,'k+')
plt.xscale('log')
plt.show()
print('Figure 5: evidence vs alpha.')
```
